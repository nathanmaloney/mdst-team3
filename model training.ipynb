```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
games_2021 = pd.read_json("../data/games_2021.json")
games_2021.columns
```

```python
games_2021.to_csv("games_2021.json")
games_2021
```

```python
big_ten = pd.read_json("../data/big_ten_2021.json")

big_ten
```

```python
df = pd.merge(games_2021, big_ten);
df
```

```python
for team in df["teams"]:
    for i in team:
        stats = i["stats"]
        for j in stats:
            feature = j["category"]
            stat_val = j["stat"]
        df[feature] = stat_val
    
```

```python
for team in df["teams"]:
    for i in team:
        school = i['school']
    df["teams"] = school
```
