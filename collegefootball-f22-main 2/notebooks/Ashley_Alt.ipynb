```python
import pandas as pd
import seaborn as sns
```

```python
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
```

```python
combined_data = games.merge(game_data, left_on='id', right_on='id', validate='one_to_one')
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
```

```python
sns.scatterplot(
    data=combined_data, x="home.kickingPoints", y="point_differential"
).set_title("Home Team Kicking Points vs Point Differential")
```

There is a positive relationship between point differential and the home team's kicking points. This means the home team won the game because they have a positive point differential.

```python
sns.scatterplot(
    data=combined_data, x="away.kickingPoints", y="point_differential"
).set_title("Away Team Total Yards vs Point Differential")
```

There is a negative relationship between point differential and the away team's kicking points. This shows that the away team lost because their point differential is negative.

```python
def total_kicking_point_diff(row):
    return row["home.kickingPoints"] - row["away.kickingPoints"]

combined_data["total_kicking_point_differential"] = combined_data.apply(total_kicking_point_diff, axis=1)
sns.scatterplot(
    data=combined_data, x="total_kicking_point_differential", y="point_differential"
).set_title("Total Kicking Point Differential vs Point Differential")
```

There is a positive relationship between total kicking points and point differential. Teams with more kicking points have a higher differential, and are more likely to win the game.
