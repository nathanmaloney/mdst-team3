```python
import pandas as pd
import seaborn as sns
```

```python
games = pd.read_csv("/Users/oviyan/Desktop/MDST/College Football Fall 2022/games.csv")
game_data = pd.read_csv("/Users/oviyan/Desktop/MDST/College Football Fall 2022/game_stats.csv")
```

```python
combined_data = games.merge(game_data, left_on='id', right_on='id', validate='one_to_one')
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
```

```python
sns.scatterplot(
    data=combined_data, x="home.rushingYards", y="point_differential"
).set_title("Home Team Rushing Yards vs Point Differential")
```

```python
sns.scatterplot(
    data=combined_data, x="away.rushingYards", y="point_differential"
).set_title("Away Team Rushing Yards vs Point Differential")
```

```python
def rushing_yard_diff(row):
    return row["home.rushingYards"] - row["away.rushingYards"]

combined_data["rushing_yard_differential"] = combined_data.apply(rushing_yard_diff, axis=1)
sns.scatterplot(
    data=combined_data, x="rushing_yard_differential", y="point_differential"
).set_title("Rushing Yard Differential vs Point Differential")
```

Based on the data as the rushing yard differential and point differential have a linear releationship. As the rush yard differential gets higher the point differential also gets higher, which shows that team with a better on the ground offense are more likely to score more during a game.
