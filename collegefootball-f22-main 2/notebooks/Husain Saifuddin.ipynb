```python
import pandas as pd
import seaborn as sns
```

```python
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
```

```python
combined_data = games.merge(game_data, left_on='id', right_on='id', validate='one_to_one')
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
```

```python
sns.scatterplot(
    data=combined_data, x="home.sacks", y="point_differential"
).set_title("Home Team Sacks vs Point Differential")
```

```python
sns.scatterplot(
    data=combined_data, x="away.sacks", y="point_differential"
).set_title("Away Team Sacks vs Point Differential")
```

```python
def total_sack_diff(row):
    return row["home.sacks"] - row["away.sacks"]

combined_data["total_sack_differential"] = combined_data.apply(total_sack_diff, axis=1)
sns.scatterplot(
    data=combined_data, x="total_sack_differential", y="point_differential"
).set_title("Total Sack Differential vs Point Differential")
```

```python
#**Overall we can see a slight trend between point differential and sacks. The more sacks generally leads to a higher point differential.**
```
