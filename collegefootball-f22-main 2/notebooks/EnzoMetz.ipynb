```python
import pandas as pd
import seaborn as sns
```

```python
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
```

```python
combined_data = games.merge(game_data, left_on='id', right_on='id', validate='one_to_one')
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
```

```python
sns.scatterplot(
    data=combined_data, x="home.totalYards", y="point_differential"
).set_title("Home Team First Downs vs Point Differential")
```

```python
sns.scatterplot(
    data=combined_data, x="away.firstDowns", y="point_differential"
).set_title("Away Team First Downs vs Point Differential")
```

```python
def total_yard_diff(row):
    return row["home.firstDowns"] - row["away.firstDowns"]

combined_data["total_yard_differential"] = combined_data.apply(total_yard_diff, axis=1)
sns.scatterplot(
    data=combined_data, x="total_yard_differential", y="point_differential"
).set_title("First Down Differential vs Point Differential")
```
