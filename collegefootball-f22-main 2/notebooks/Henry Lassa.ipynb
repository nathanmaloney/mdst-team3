```python
#Henry Lassa

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

games = pd.read_csv("games.csv")
game_stats = pd.read_csv("game_stats.csv")

#games.head()
#list(games)
#games

combined_data = games.merge(game_stats, left_on='id', right_on='id', validate='one_to_one')
combined_data
```

```python
#df.dropna(subset=['AVG_READING_8_SCORE'], inplace=True)
cleaned = combined_data.dropna(subset=['home_points', 'away_points'])
cleaned
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

cleaned["point_differential"] = cleaned.apply(point_diff, axis=1)
```

```python
sb.scatterplot(
    data=cleaned, x="home.totalYards", y="point_differential"
).set_title("Home Team Total Yards vs Point Differential")
```

```python
#This is the example given in the GitHub repo. If the home team scores more total yards, 
#they are more likely to have a higher point differential
```

```python
sb.scatterplot(
    data=cleaned, x="away.rushingAttempts", y="point_differential"
).set_title("Home Team Total Yards vs Point Differential")
sb.scatterplot(
    data=cleaned, x="home.rushingAttempts", y="point_differential"
).set_title("Home Team Total Yards vs Point Differential")
```

```python
#This shows the relationship between both the home and away teams for rushing
#attempts vs point differential. We can see a positive correlation between the 
# home team's rushing attempts and point differential, and a negative correlation
# for the away teams'.
```
