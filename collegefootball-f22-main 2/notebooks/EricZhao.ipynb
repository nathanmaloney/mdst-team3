```python
import numpy as np
import pandas as pd
import seaborn as sns

# import and read the file:
gameStats = pd.read_csv("../data/game_stats.csv")
game = pd.read_csv("../data/games.csv")
```

```python
# combine both csv files

combinedData = game.merge(gameStats, left_on='id', right_on='id', validate='one_to_one')
```

```python
# function to create point differential column:

def pointDiff(row):
    return row["home_points"] - row["away_points"]

combinedData["point_differential"] = combinedData.apply(pointDiff, axis=1)
```

```python
# create scatterplot- reference seaborn library

sns.scatterplot(
    data=combinedData, x="home.firstDowns", y="point_differential"
).set_title("")
```

There is a loosely proportional relationship between first downs and a positive point differential; the more first downs our team has, the liklier we are to win a game.
