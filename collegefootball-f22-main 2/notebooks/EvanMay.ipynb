```python
import pandas as pd 
import seaborn as sns 
```

```python
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
```

```python
combined_data = games.merge(game_data, left_on='id', right_on='id', validate='one_to_one')
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
```

```python
def netPassingYards_diff(row):
    return row["home.netPassingYards"] - row["away.netPassingYards"]

combined_data["netPassingYards_diff"] = combined_data.apply(netPassingYards_diff, axis=1)
sns.scatterplot(
    data=combined_data, x="netPassingYards_diff", y="point_differential"
).set_title("Turnover Differential vs Point Differential")
```

This graph shows the difference in net passing yards between the home and away teams and compares those values to the point differential of the game. Overall there is a slight possitive correlation between a team having a higher amount of net passing yards and winning the game.
