```python

import pandas as pd
```

```python
games_data = pd.read_csv("games.csv")
```

```python
games_stats_data = pd.read_csv("game_stats.csv")
```

```python
games_stats_data.head()
```

```python
games_data.head()
```

```python
games_data["point_differential"] = games_data["home_points"] - games_data["away_points"]
games_data.head()
```

```python
data = games_data.merge(games_stats_data, on=["id"])
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plotHome = data.plot.scatter(x='home.yardsPerRushAttempt', y='point_differential')
```

```python
plotAway = data.plot.scatter(x='away.yardsPerRushAttempt', y='point_differential')
```
