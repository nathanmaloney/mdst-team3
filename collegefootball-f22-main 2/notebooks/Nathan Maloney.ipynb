```python
import pandas as pd
import seaborn as sns
```

```python
game_data = pd.read_csv("../data/game_stats.csv")
games = pd.read_csv("../data/games.csv")
```

```python
combined_data = games.merge(game_data, left_on='id', right_on='id', validate='one_to_one')
```

```python
def point_diff(row):
    return row["home_points"] - row["away_points"]

combined_data["point_differential"] = combined_data.apply(point_diff, axis=1)
```

```python
sns.scatterplot(
    data=combined_data, x="home.firstDowns", y="point_differential"
).set_title("Home Team First Downs vs Point Differential")
```

```python
def diff_in_first_downs(row):
    return row["home.firstDowns"] - row["away.firstDowns"]

combined_data["diff_in_first_downs"] = combined_data.apply(diff_in_first_downs, axis=1)
sns.scatterplot(
    data=combined_data, x="diff_in_first_downs", y="point_differential"
).set_title("Difference in Amount of First Downs vs Point Differential")
```
